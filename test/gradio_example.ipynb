# gradio 사용법 익히기
[여기](https://www.gradio.app/guides/quickstart)의 예제부터 시작

```python
import gradio as gr

def greet(name, intensity):
    return "Hello, " + name + "!" * int(intensity)

demo = gr.Interface(
    fn=greet,
    inputs=["text", "slider"],
    outputs=["text"],
    api_name="predict"
)

demo.launch()
# demo.launch(share=True) # 보안이 이걸 막는데?
```
